```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import cosine_similarity
import plotly.graph_objects as go
```

```python
df = pd.read_csv('raw_data.csv', encoding = "UTF-8")
df.head(11)
```

```python
df.shape
```

```python
idx = []
for i in range(1,len(df)):
    if df.at[i,'Player'] == df.at[(i-1),'Player']:
        idx.append(i)
df = df.drop(idx)
```

```python
df.shape
```

```python
df = df.reset_index(drop=True)
```

```python
ids_ = df[['Player', 'year']]
ids_.head()
```

```python
df = df.drop(['Player', 'Tm', 'year'], axis = 1)
df.head()
```

```python
df = pd.get_dummies(df, columns = ['Pos'])
df.head()
```

```python
df.shape
```

```python
cs = cosine_similarity(df)
```

```python
def find_comps(player, cs = cs):
    comps = []
    # get the index from our series of titles
    posses = ids_[ids_['Player'] == player]
    idx = posses[posses['year']==2019].index[0]
    # creating a Series of descending similarity scores
    scores = pd.Series(cs[idx]).sort_values(ascending = False)
    # making a list of the top six scores
    top_5 = list(scores.iloc[1:7].index)
    # populating the list with the corresponding movie titles
    for i in top_5:
        if ids_.loc[i][0] != player:
            comps.append((ids_.loc[i][0], ids_.loc[i][1]))
    return comps
```
