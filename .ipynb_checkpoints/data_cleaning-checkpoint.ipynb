```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.decomposition import PCA
import plotly.graph_objects as go
```

```python
df = pd.read_csv('raw_data.csv')
```

```python
df.head(11)
```

```python
idx = []
for i in range(1,len(df)):
    if df.at[i,'Player'] == df.at[(i-1),'Player']:
        idx.append(i)
```

```python
df = df.drop(idx)
```

```python
df.shape
```

```python
player_ids = {ident: (player, year) for ident, player, year in zip(list(df.index), list(df['Player']), list(df['year']))}
```

```python
df = pd.get_dummies(df, columns=['Pos'])
```

```python
X = df[df['year'] == 2019]
X = df.drop(['Player', 'Tm', 'year'], axis=1).values
```

```python
X = StandardScaler().fit_transform(X)
```

```python
pca = PCA(n_components = 2)
principalComponents = pca.fit_transform(X)
```

```python
pdf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
pdf.head()
```

```python
df[['Player']].head()
```

```python
finalDf = pd.concat([pdf, df[['Player']]], axis = 1)
finalDf.head(5)
```

```python
fig = go.Figure(go.Scatter(x=finalDf['principal component 1'], y=finalDf['principal component 2'], name = list(finalDf['Player'])))
fig.show()
```

```python
kmeans = kmeans = KMeans(n_clusters=115, max_iter=600, algorithm = 'auto')
kmeans.fit(X)
```
